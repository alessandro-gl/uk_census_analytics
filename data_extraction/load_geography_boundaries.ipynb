```python
from google.cloud import bigquery
import geopandas as gpd
import pandas as pd

# Read the GeoJSON file
gdf = gpd.read_file('Westminster_Parliamentary_Constituencies_Dec_2021_UK_BGC_2022_-5235546284095580872.geojson')

# Convert to WKT (Well-Known Text) for BigQuery
gdf['geometry_wkt'] = gdf['geometry'].apply(lambda x: x.wkt)

# Select relevant columns
df = pd.DataFrame({
    'pcon21cd': gdf['PCON21CD'],
    'pcon21nm': gdf['PCON21NM'],
    'geometry_wkt': gdf['geometry_wkt']
})

# Upload to BigQuery
df.to_gbq(
    destination_table='raw_data.geography_boundaries_2021',
    project_id='uk-census-portfolio',  # Replace with your project ID
    if_exists='replace',
    progress_bar=True
)

print("Upload complete!")
```
